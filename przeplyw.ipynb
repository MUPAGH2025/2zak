```python
import pint
si=pint.UnitRegistry()
```

```python
2*si.kg+2*si.mg
```

```python
from scipy import constants
Rstr=constants.R*si.J/si.K/si.mol
Md=28.966*si.kg/si.kmol
R=Rstr/Md
def rho(p,T):
    return p/R/T
```

```python
rho(1000*si.hPa,300*si.K).to_base_units()
```

```python
from matplotlib import pyplot
import numpy as np
si.setup_matplotlib()
```

```python
pyplot.plot(
    T:=np.linspace(280,320)*si.K,
    rho(1000*si.hPa,T).to_base_units(),label="gęstość powietrza")
pyplot.gca().xaxis.set_units(si.celsius)
show_plot()
```
